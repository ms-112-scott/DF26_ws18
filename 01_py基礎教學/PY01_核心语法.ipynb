## 变量、类型与注解

变量 variable 是一个「名称」，绑定到内存中的某个值；赋值 assignment 用 `=` 把右边的值绑到左边的名称。

为什么学：所有程序都靠变量暂存数据、传递结果，这是后续一切语法的地基。

四种基本类型：整数 int、浮点数 float、字符串 str、布尔 bool。Python 会依值自动判断类型，不必事先声明。

注解 comment 以 `#` 开头，直到行尾；缩进 indentation（通常 4 空格）用来界定程序区块，不是排版装饰。

````python
# 概念：变量赋值与四种基本类型
# 造一笔假个人数据当示范用，每个变量对应一种类型
name = "林小明"        # str：用引号包住的文字
age = 21               # int：整数，无小数点
height = 173.5         # float：浮点数，有小数点
is_student = True      # bool：布尔值，只有 True / False

# type() 会回传变量目前绑定值的类型，用来确认自动推断的结果
print(name, type(name))
print(age, type(age))
print(height, type(height))
print(is_student, type(is_student))

# 注意：同一个名称可重新赋值成不同类型，Python 不会报错（动态类型 dynamic typing）
age = "二十一"          # 原本是 int，现在改绑到 str
print("重新赋值后：", age, type(age))
````

## 操作符与表达式

操作符 operator 把值组合成表达式 expression，表达式会求出一个结果值。

为什么学：计算与条件判断都靠操作符；理解它们的回传类型，才能正确串接逻辑。

| 类别 | 操作符 | 回传 |
|---|---|---|
| 算术 arithmetic | `+ - * / // % **` | 数值 |
| 比较 comparison | `== != > < >= <=` | 布尔 bool |
| 逻辑 logical | `and` `or` `not` | 布尔 bool |

比较运算的结果是布尔值，逻辑操作符再把多个布尔值串接成一个整体条件。

````python
# 概念：用算术算出 BMI，再用比较与逻辑判断是否落在正常区间
# 自造一笔身高体重数据当示范用
height_m = 1.735       # 公尺
weight_kg = 64.0       # 公斤

bmi = weight_kg / height_m ** 2   # ** 是次方，先算次方再除（操作符有优先序）
print("BMI =", bmi)

# 比较运算回传布尔值
print("BMI >= 18.5 吗？", bmi >= 18.5)

# 技巧：Python 允许连续比较 18.5 <= bmi < 24，等同 (bmi >= 18.5) and (bmi < 24)
is_normal = 18.5 <= bmi < 24      # 一次表达「介于两值之间」
print("落在正常区间吗？", is_normal)

# 逻辑操作符串接多个条件
print("过轻或过重？", bmi < 18.5 or bmi >= 24)
````

## 条件判断

`if / elif / else` 依条件真假决定运行哪一段程序，把现实的判断拆成多个分支。

为什么学：程序要「依情况做不同事」就靠条件分支。

真值测试 truthiness：条件不一定要写成明确比较，对象本身就有真假值。空容器、`0`、空字符串、`None` 视为假 falsy，其余多为真 truthy。

形状：
```
if 条件:
    ...
elif 另一条件:
    ...
else:
    ...
```

````python
# 概念：依分数判断等第，并用真值测试判断输入是否为空
score = 76             # 自造一个分数当示范用

if score >= 90:
    grade = "A"
elif score >= 80:      # 走到这里代表前面条件已为假
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "F"
print("分数", score, "等第", grade)

# 概念：真值测试 truthiness，直接以对象真假当条件，不必写 == 比较
user_input = ""        # 空字符串视为假
if user_input:                       # 等同 if len(user_input) > 0
    print("收到输入：", user_input)
else:
    print("输入为空，请重新输入")

# 注意：0、0.0、None、空 list / dict 也都视为假，别误用 == True 比较
for value in [0, None, [], "x", 5]:
    print(repr(value), "->", bool(value))
````

## for 循环与 range()

`for` 循环 for loop 走访一个串行，对每个元素重复运行同一段程序。

为什么学：要「重复固定次数」或「逐一处理每笔数据」时就用 for。

`range(start, stop, step)` 产生一段整数串行（不含 stop），是计次循环最常见的来源。配合累加器 accumulator（先设 0 的变量，循环内逐步加总）可完成求和、计数等任务。

形状：
```
for 变量 in 可迭代物:
    ...
```

````python
# 概念：用 range() 走访数字，搭配累加器求总和与平均
total = 0              # 累加器 accumulator，先归零
count = 0              # 计数器，记录走访了几个数

# range(1, 11) 产生 1 到 10（不含 11）
for n in range(1, 11):
    total = total + n  # 把当前数字累加进 total
    count = count + 1

average = total / count
print("总和 =", total)
print("个数 =", count)
print("平均 =", average)

# 技巧：要同时拿到索引与元素时用 enumerate，比手动维护计数器清楚
heights = [158, 172, 165]          # 自造一组身高（公分）
for i, h in enumerate(heights):
    print("第", i, "笔身高", h)
````

## while 循环与流程跳脱

`while` 循环 while loop 只要条件成立就重复运行，适合「不知道要跑几次、跑到满足某条件为止」的情况。

为什么学：收敛、轮询、交互等次数不固定的流程都靠 while。

三个流程控制关键字：`break` 立刻跳出整个循环、`continue` 跳过本轮剩余程序直接进入下一轮、`pass` 什么都不做（占位）。

二分逼近 binary-search convergence：在一个区间内反复取中点、依结果缩小区间，逐步逼近答案。

````python
# 概念：用二分逼近法求平方根，示范 while 收敛与流程跳脱
target = 2.0           # 想求 target 的平方根，理论值约 1.41421
low, high = 0.0, target            # 初始区间 [low, high]
tolerance = 1e-6       # 容许误差，够小就停
steps = 0

while True:            # 先进循环，内部再用 break 决定何时停
    mid = (low + high) / 2         # 取区间中点当猜测值
    error = mid * mid - target     # 中点平方与目标的差距

    # 注意：误差用绝对值判断，正负都算
    if abs(error) < tolerance:
        break          # 够接近就跳出整个循环

    if error > 0:      # 中点平方偏大，答案在左半边
        high = mid
    else:              # 偏小，答案在右半边
        low = mid
    steps = steps + 1

print("逼近结果 =", mid)
print("迭代次数 =", steps)

# 技巧：continue 与 pass 示范 —— 只印出 1 到 5 的奇数
for k in range(1, 6):
    if k % 2 == 0:
        continue       # 偶数：跳过本轮剩余程序
    else:
        pass           # pass 占位：此分支暂时不做事，仅示范语法
    print("奇数", k)
````

## 输出与 f-string 格式化

`print()` 可接收多个参数 multiple arguments，会以空白分隔依序印出；f-string 则把变量直接嵌进字符串。

为什么学：呈现结果是程序的最后一哩路，整齐的输出让数据一目了然。

f-string 在字符串前加 `f`，用 `{变量}` 嵌入值，并可加格式规格 format specification 控制呈现：

| 格式 | 意义 |
|---|---|
| `:.4f` | 浮点数，固定 4 位小数 |
| `:>8` | 靠右对齐，宽度 8 |
| `:,` | 千分位逗号 |

````python
# 概念：用 f-string 与格式规格把数值整齐打印成一行报表
average = 5.5          # 沿用前面求和的平均当示范值
sqrt_two = 1.4142135623730951      # 平方根逼近结果

# print 多参数：以逗号分隔，输出时自动补空白
print("平均", average, "平方根", sqrt_two)

# f-string：{} 内可直接放变量，冒号后是格式规格
print(f"平均 = {average:.2f}，根号 2 = {sqrt_two:.4f}")

# 技巧：对齐与千分位让报表更易读
gfa = 3840.0
print(f"总楼地板面积 = {gfa:>10,.1f} m²")    # 靠右、宽 10、千分位、1 位小数
````

## 字符串方法与旧式格式化

字符串方法 string method 是字符串对象自带的函数，用来拆解、清理与重组文字。

为什么学：真实数据常以文字进来（CSV 行、表单字段），要先清理才能使用。

| 方法 | 作用 |
|---|---|
| `split(sep)` | 依分隔符切成 list |
| `strip()` | 去除头尾空白 |
| `join(seq)` | 用指定分隔符把串行接成字符串 |
| `replace(a, b)` | 把 a 换成 b |
| `lower()` | 转小写 |
| `startswith(s)` | 是否以 s 开头 |

另外两种旧式格式化 old-style formatting：`"{}".format(...)` 与 `"%s" % (...)`，在既有代码中仍会遇到。

````python
# 概念：把一行逗号分隔字符串切开、清理、转小写后重新组合
raw = "  Apple , Banana , Cherry ,Apple "   # 自造一行脏数据，含多余空白

parts = raw.split(",")             # 依逗号切成 list
print("切开后：", parts)

# 对每个片段去头尾空白并转小写（串列推导式之后章节会深入，这里先用 for）
cleaned = []
for p in parts:
    cleaned.append(p.strip().lower())   # strip 去空白后接着 lower 转小写
print("清理后：", cleaned)

# join 把 list 接回字符串，分隔符是调用 join 的那个字符串
result = " | ".join(cleaned)
print("重组后：", result)

# replace 与 startswith 示范
print("取代：", result.replace("apple", "AAA"))
print("以 apple 开头？", cleaned[0].startswith("apple"))

# 概念：旧式格式化 old-style formatting，认得即可，新程序优先用 f-string
name, count = "apple", 2
print("{} 出现 {} 次".format(name, count))   # str.format 写法
print("%s 出现 %d 次" % (name, count))         # % 写法（%s 字符串、%d 整数）
````
