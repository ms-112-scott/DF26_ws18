## 變數、型別與註解

變數 variable 是一個「名稱」，綁定到記憶體中的某個值；賦值 assignment 用 `=` 把右邊的值綁到左邊的名稱。

為什麼學：所有程式都靠變數暫存資料、傳遞結果，這是後續一切語法的地基。

四種基本型別：整數 int、浮點數 float、字串 str、布林 bool。Python 會依值自動判斷型別，不必事先宣告。

註解 comment 以 `#` 開頭，直到行尾；縮排 indentation（通常 4 空格）用來界定程式區塊，不是排版裝飾。

````python
# 概念：變數賦值與四種基本型別
# 造一筆假個人資料當示範用，每個變數對應一種型別
name = "林小明"        # str：用引號包住的文字
age = 21               # int：整數，無小數點
height = 173.5         # float：浮點數，有小數點
is_student = True      # bool：布林值，只有 True / False

# type() 會回傳變數目前綁定值的型別，用來確認自動推斷的結果
print(name, type(name))
print(age, type(age))
print(height, type(height))
print(is_student, type(is_student))

# 注意：同一個名稱可重新賦值成不同型別，Python 不會報錯（動態型別 dynamic typing）
age = "二十一"          # 原本是 int，現在改綁到 str
print("重新賦值後：", age, type(age))
````

## 運算子與運算式

運算子 operator 把值組合成運算式 expression，運算式會求出一個結果值。

為什麼學：計算與條件判斷都靠運算子；理解它們的回傳型別，才能正確串接邏輯。

| 類別 | 運算子 | 回傳 |
|---|---|---|
| 算術 arithmetic | `+ - * / // % **` | 數值 |
| 比較 comparison | `== != > < >= <=` | 布林 bool |
| 邏輯 logical | `and` `or` `not` | 布林 bool |

比較運算的結果是布林值，邏輯運算子再把多個布林值串接成一個整體條件。

````python
# 概念：用算術算出 BMI，再用比較與邏輯判斷是否落在正常區間
# 自造一筆身高體重資料當示範用
height_m = 1.735       # 公尺
weight_kg = 64.0       # 公斤

bmi = weight_kg / height_m ** 2   # ** 是次方，先算次方再除（運算子有優先序）
print("BMI =", bmi)

# 比較運算回傳布林值
print("BMI >= 18.5 嗎？", bmi >= 18.5)

# 技巧：Python 允許連續比較 18.5 <= bmi < 24，等同 (bmi >= 18.5) and (bmi < 24)
is_normal = 18.5 <= bmi < 24      # 一次表達「介於兩值之間」
print("落在正常區間嗎？", is_normal)

# 邏輯運算子串接多個條件
print("過輕或過重？", bmi < 18.5 or bmi >= 24)
````

## 條件判斷

`if / elif / else` 依條件真假決定執行哪一段程式，把現實的判斷拆成多個分支。

為什麼學：程式要「依情況做不同事」就靠條件分支。

真值測試 truthiness：條件不一定要寫成明確比較，物件本身就有真假值。空容器、`0`、空字串、`None` 視為假 falsy，其餘多為真 truthy。

形狀：
```
if 條件:
    ...
elif 另一條件:
    ...
else:
    ...
```

````python
# 概念：依分數判斷等第，並用真值測試判斷輸入是否為空
score = 76             # 自造一個分數當示範用

if score >= 90:
    grade = "A"
elif score >= 80:      # 走到這裡代表前面條件已為假
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "F"
print("分數", score, "等第", grade)

# 概念：真值測試 truthiness，直接以物件真假當條件，不必寫 == 比較
user_input = ""        # 空字串視為假
if user_input:                       # 等同 if len(user_input) > 0
    print("收到輸入：", user_input)
else:
    print("輸入為空，請重新輸入")

# 注意：0、0.0、None、空 list / dict 也都視為假，別誤用 == True 比較
for value in [0, None, [], "x", 5]:
    print(repr(value), "->", bool(value))
````

## for 迴圈與 range()

`for` 迴圈 for loop 走訪一個序列，對每個元素重複執行同一段程式。

為什麼學：要「重複固定次數」或「逐一處理每筆資料」時就用 for。

`range(start, stop, step)` 產生一段整數序列（不含 stop），是計次迴圈最常見的來源。配合累加器 accumulator（先設 0 的變數，迴圈內逐步加總）可完成求和、計數等任務。

形狀：
```
for 變數 in 可迭代物:
    ...
```

````python
# 概念：用 range() 走訪數字，搭配累加器求總和與平均
total = 0              # 累加器 accumulator，先歸零
count = 0              # 計數器，記錄走訪了幾個數

# range(1, 11) 產生 1 到 10（不含 11）
for n in range(1, 11):
    total = total + n  # 把當前數字累加進 total
    count = count + 1

average = total / count
print("總和 =", total)
print("個數 =", count)
print("平均 =", average)

# 技巧：要同時拿到索引與元素時用 enumerate，比手動維護計數器清楚
heights = [158, 172, 165]          # 自造一組身高（公分）
for i, h in enumerate(heights):
    print("第", i, "筆身高", h)
````

## while 迴圈與流程跳脫

`while` 迴圈 while loop 只要條件成立就重複執行，適合「不知道要跑幾次、跑到滿足某條件為止」的情況。

為什麼學：收斂、輪詢、互動等次數不固定的流程都靠 while。

三個流程控制關鍵字：`break` 立刻跳出整個迴圈、`continue` 跳過本輪剩餘程式直接進入下一輪、`pass` 什麼都不做（佔位）。

二分逼近 binary-search convergence：在一個區間內反覆取中點、依結果縮小區間，逐步逼近答案。

````python
# 概念：用二分逼近法求平方根，示範 while 收斂與流程跳脫
target = 2.0           # 想求 target 的平方根，理論值約 1.41421
low, high = 0.0, target            # 初始區間 [low, high]
tolerance = 1e-6       # 容許誤差，夠小就停
steps = 0

while True:            # 先進迴圈，內部再用 break 決定何時停
    mid = (low + high) / 2         # 取區間中點當猜測值
    error = mid * mid - target     # 中點平方與目標的差距

    # 注意：誤差用絕對值判斷，正負都算
    if abs(error) < tolerance:
        break          # 夠接近就跳出整個迴圈

    if error > 0:      # 中點平方偏大，答案在左半邊
        high = mid
    else:              # 偏小，答案在右半邊
        low = mid
    steps = steps + 1

print("逼近結果 =", mid)
print("迭代次數 =", steps)

# 技巧：continue 與 pass 示範 —— 只印出 1 到 5 的奇數
for k in range(1, 6):
    if k % 2 == 0:
        continue       # 偶數：跳過本輪剩餘程式
    else:
        pass           # pass 佔位：此分支暫時不做事，僅示範語法
    print("奇數", k)
````

## 輸出與 f-string 格式化

`print()` 可接收多個參數 multiple arguments，會以空白分隔依序印出；f-string 則把變數直接嵌進字串。

為什麼學：呈現結果是程式的最後一哩路，整齊的輸出讓資料一目了然。

f-string 在字串前加 `f`，用 `{變數}` 嵌入值，並可加格式規格 format specification 控制呈現：

| 格式 | 意義 |
|---|---|
| `:.4f` | 浮點數，固定 4 位小數 |
| `:>8` | 靠右對齊，寬度 8 |
| `:,` | 千分位逗號 |

````python
# 概念：用 f-string 與格式規格把數值整齊列印成一行報表
average = 5.5          # 沿用前面求和的平均當示範值
sqrt_two = 1.4142135623730951      # 平方根逼近結果

# print 多參數：以逗號分隔，輸出時自動補空白
print("平均", average, "平方根", sqrt_two)

# f-string：{} 內可直接放變數，冒號後是格式規格
print(f"平均 = {average:.2f}，根號 2 = {sqrt_two:.4f}")

# 技巧：對齊與千分位讓報表更易讀
gfa = 3840.0
print(f"總樓地板面積 = {gfa:>10,.1f} m²")    # 靠右、寬 10、千分位、1 位小數
````

## 字串方法與舊式格式化

字串方法 string method 是字串物件自帶的函式，用來拆解、清理與重組文字。

為什麼學：真實資料常以文字進來（CSV 行、表單欄位），要先清理才能使用。

| 方法 | 作用 |
|---|---|
| `split(sep)` | 依分隔符切成 list |
| `strip()` | 去除頭尾空白 |
| `join(seq)` | 用指定分隔符把序列接成字串 |
| `replace(a, b)` | 把 a 換成 b |
| `lower()` | 轉小寫 |
| `startswith(s)` | 是否以 s 開頭 |

另外兩種舊式格式化 old-style formatting：`"{}".format(...)` 與 `"%s" % (...)`，在既有程式碼中仍會遇到。

````python
# 概念：把一行逗號分隔字串切開、清理、轉小寫後重新組合
raw = "  Apple , Banana , Cherry ,Apple "   # 自造一行髒資料，含多餘空白

parts = raw.split(",")             # 依逗號切成 list
print("切開後：", parts)

# 對每個片段去頭尾空白並轉小寫（串列推導式之後章節會深入，這裡先用 for）
cleaned = []
for p in parts:
    cleaned.append(p.strip().lower())   # strip 去空白後接著 lower 轉小寫
print("清理後：", cleaned)

# join 把 list 接回字串，分隔符是呼叫 join 的那個字串
result = " | ".join(cleaned)
print("重組後：", result)

# replace 與 startswith 示範
print("取代：", result.replace("apple", "AAA"))
print("以 apple 開頭？", cleaned[0].startswith("apple"))

# 概念：舊式格式化 old-style formatting，認得即可，新程式優先用 f-string
name, count = "apple", 2
print("{} 出現 {} 次".format(name, count))   # str.format 寫法
print("%s 出現 %d 次" % (name, count))         # % 寫法（%s 字串、%d 整數）
````

## 小結
