用類別 class 把「資料」與「操作資料的行為」綁在一起，學會以物件思維組織程式。

## 從字典到類別：為什麼需要 class

類別（class）是一張「資料藍圖」，把同一種事物的資料與行為綁在一起。

新手常先用字典（dict）裝一筆資料，例如一個點的座標。dict 能裝資料，但「對資料做的操作」（如算距離）只能寫成散落各處的獨立函式，資料與行為各走各的。

當這種資料會被重複使用、又需要附帶行為時，就該把資料與行為封裝（encapsulation）在一起，這正是 class 的動機。class 定義出型別，依藍圖造出的個別物件叫實例（instance）。

````python
# 概念：用 dict 裝一筆資料 vs 用 class 把資料與行為綁在一起

# 造一筆假的點座標資料當示範用
point_dict = {'x': 3, 'y': 4}                 # dict 能裝資料，但行為要另外寫

# 注意：算距離得寫成獨立函式，資料與行為散落兩處
def distance_from_origin(p):
    return (p['x'] ** 2 + p['y'] ** 2) ** 0.5

print('dict 寫法：', distance_from_origin(point_dict))

# 改寫成 class 雛形：資料（x, y）與行為（dist）綁在同一個藍圖
class Point:
    def __init__(self, x, y):                 # 建立物件當下設定座標
        self.x = x
        self.y = y
    def dist(self):                           # 行為就住在物件身上
        return (self.x ** 2 + self.y ** 2) ** 0.5

p = Point(3, 4)                               # 依藍圖造出一個實例
print('class 寫法：', p.dist())
````

## 定義類別：__init__ 與 self

`__init__` 是建構子（constructor），在建立物件的當下自動執行，負責設定物件的初始狀態。

`self` 代表「這個物件自己」，是每個方法的第一個參數（呼叫時不用自己傳）。把值存到 `self.屬性名`，就成了這個實例專屬的實例屬性（instance attribute）。

形狀：
```
class 類別名:
    def __init__(self, 參數...):
        self.屬性 = 參數
```
用 `類別名(參數...)` 建立實例（instantiation）。

````python
# 概念：用 __init__ 與 self 初始化實例屬性
class Rectangle:
    def __init__(self, width, height):        # 建立物件時自動呼叫
        self.width = width                    # 把寬存成這個物件自己的屬性
        self.height = height                  # 把高存成這個物件自己的屬性

# 注意：呼叫時不用自己傳 self，Python 會把物件本身塞進去
r1 = Rectangle(4, 3)                          # 造第一個矩形
r2 = Rectangle(10, 2)                         # 造第二個，各自獨立

# 技巧：用點記號讀回每個實例自己的屬性
print('r1：', r1.width, 'x', r1.height)
print('r2：', r2.width, 'x', r2.height)
````

## 實例方法與屬性存取

實例方法（instance method）是「屬於物件的函式」，定義在 class 內、第一個參數為 `self`，因此能直接讀寫自己的屬性。

它和一般函式的差別在於：方法掛在物件身上，透過點記號（dot notation）`物件.方法()` 呼叫，操作的就是該物件的資料。方法可以用 `return` 回傳計算結果，也可以就地修改屬性，讓物件不只裝資料、還會做事。

````python
# 概念：實例方法讀寫自己的屬性（area 計算、scale 就地修改）
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):                           # 讀屬性算面積後回傳
        return self.width * self.height
    def scale(self, factor):                  # 就地放大尺寸（不回傳）
        self.width = self.width * factor
        self.height = self.height * factor

r = Rectangle(4, 3)
print('原始面積：', r.area())                 # 透過點記號呼叫方法

r.scale(2)                                    # 改變物件內部狀態
print('放大後尺寸：', r.width, 'x', r.height)
print('放大後面積：', r.area())
````

## 方法鏈 method chaining

方法鏈（method chaining）是讓方法回傳 `self`，使多個操作能在一行內串接呼叫。

因為回傳的是物件自己，下一個點記號就接著對同一個可變物件（mutable object）操作。這種一路 `.add().upper()` 接下去的寫法稱為流暢介面（fluent interface），好處是讀起來像一連串動作；取捨是串太長時不易除錯，需自行拿捏可讀性。

形狀：
```
def 方法(self, ...):
    ...修改 self...
    return self
```

````python
# 概念：方法回傳 self 以支援方法鏈
class TextBuilder:
    def __init__(self):
        self.text = ''                        # 初始為空字串
    def add(self, s):
        self.text = self.text + s
        return self                           # 回傳自己，才能繼續往下接
    def upper(self):
        self.text = self.text.upper()
        return self

# 技巧：一行串接多個操作，像在描述一連串動作
result = TextBuilder().add('hello ').add('world').upper().text
print(result)
````

## 繼承 inheritance 與 super()

繼承（inheritance）讓子類別（subclass）沿用父類別（parent class）的屬性與方法，只補上差異，避免重複程式碼。

語法是 `class 子類(父類):`。子類別若也有自己的 `__init__`，要用 `super().__init__(...)` 先把父類別該初始化的部分做好，再加上自己新增的屬性，否則父類別設定的屬性會缺漏。

形狀：
```
class 子類(父類):
    def __init__(self, ...):
        super().__init__(...)
        self.新屬性 = ...
```

````python
# 概念：子類別用 super().__init__() 初始化父類別部分，再補自己的屬性
class Shape:
    def __init__(self, name):
        self.name = name                      # 父類別負責共通屬性

class Circle(Shape):                          # Circle 繼承自 Shape
    def __init__(self, radius):
        super().__init__('circle')            # 先讓父類別設好 name
        self.radius = radius                  # 再補上子類別專屬屬性
    def area(self):
        return 3.14159 * self.radius ** 2

c = Circle(5)
# 注意：name 來自父類別，radius 來自子類別，兩者都在同一個物件上
print('名稱（繼承自父類別）：', c.name)
print('面積：', c.area())
````

## 方法覆寫與慣例方法

方法覆寫（method overriding）是子類別定義一個與父類別同名的方法，用自己的版本取代父類別的行為。

當多個子類別都提供同名方法（共同介面 common interface），就能用同一個方法名統一呼叫、各自跑出不同結果，這就是多型（polymorphism）的直覺。實務上常約定一個固定方法名（例如機器學習框架的 `forward`）作為各子類別的共同入口。

````python
# 概念：多個子類別覆寫同名方法 area，用統一介面呼叫（多型）
class Shape:
    def area(self):                           # 父類別給出共同介面
        return 0

class Square(Shape):
    def __init__(self, side):
        self.side = side
    def area(self):                           # 覆寫成正方形的算法
        return self.side ** 2

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):                           # 覆寫成圓形的算法
        return 3.14159 * self.radius ** 2

# 技巧：不同子類別放同一清單，用同一個方法名呼叫即可
shapes = [Square(4), Circle(3), Square(2)]
for s in shapes:
    print(type(s).__name__, '面積：', s.area())
````

## dict 還是 class：設計取捨

字典（dict）與類別（class）都能當資料容器（data container），選哪個是設計取捨。

判斷準則：

| 情況 | 建議 |
|---|---|
| 臨時、鬆散、欄位不固定的資料 | 用 dict |
| 需要附帶行為（方法） | 用 class |
| 結構固定、會被重複使用 | 用 class |
| 只是一次性傳遞少量鍵值 | 用 dict |

為簡單資料硬寫一堆 class 是過度設計（over-engineering），反而傷可讀性與可維護性；兩端都要避免極端。

````python
# 概念：同一筆學生成績，用 dict 與 class 各表達一次，比較差異

# 用 dict：快、輕，但行為（算平均）得另寫函式
student_dict = {'name': 'Amy', 'scores': [80, 90, 70]}
avg_dict = sum(student_dict['scores']) / len(student_dict['scores'])
print('dict 版平均：', avg_dict)

# 用 class：結構固定、行為內建，會重複使用時更清楚
class Student:
    def __init__(self, name, scores):
        self.name = name
        self.scores = scores
    def average(self):                        # 行為就住在物件身上
        return sum(self.scores) / len(self.scores)

s = Student('Amy', [80, 90, 70])
print('class 版平均：', s.average())
# 注意：臨時用一次選 dict，固定結構又有行為選 class
````

## 小結

- 類別（class）是資料藍圖，把資料與行為封裝在一起；依藍圖造出的個別物件是實例（instance）。
- `__init__` 在建立物件時自動執行，`self` 代表物件自己，用 `self.屬性` 設定實例屬性。
- 實例方法掛在物件身上，透過點記號讀寫自己的屬性；讓方法回傳 `self` 即可串成方法鏈。
- 繼承（inheritance）讓子類別沿用父類別並補上差異，子類別的 `__init__` 要先呼叫 `super().__init__()`。
- 方法覆寫（method overriding）配合共同介面帶來多型（polymorphism）：同一個方法名、各子類別各自的行為。
- 鬆散臨時資料用 dict，固定結構又需要行為時用 class，避免過度設計。
