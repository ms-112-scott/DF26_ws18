## 清单与索引切片

清单 list 是一串有顺序、可以随时修改的数据，是 Python 最常用的可变串行 mutable sequence。

先打稳元素的添加、定位与区段截取，后续的推导式才有着力点。常用操作：

| 操作 | 写法 | 作用 |
|---|---|---|
| 附加 append | `xs.append(v)` | 在尾端加入单一元素 |
| 延伸 extend | `xs.extend(ys)` | 把另一串元素逐一接到尾端 |
| 索引 indexing | `xs[i]` | 取出第 i 个元素（从 0 起算） |
| 切片 slicing | `xs[a:b]` | 取出 a 到 b（不含 b）的区段 |

````python
# 概念：清单创建、append、extend、索引与切片
# 造一周的每日气温（摄氏度）当示范数据
temps = [22, 24, 19, 21, 25, 23, 20]

temps.append(26)              # 在尾端补上「第八天」一笔新值
temps.extend([18, 17])        # 一次接上两天数据（extend 是逐一展开，不是塞入一个 list）

# 注意：索引从 0 起算，负索引从尾端倒数，-1 是最后一个
print("第一天：", temps[0])
print("最后一天：", temps[-1])

# 切片 [a:b] 取 a 到 b-1；省略起点代表从头，负数从尾端算
print("后三天：", temps[-3:])   # 取最后三笔
print("全部：", temps)
````

## 字典的键值操作

字典 dict 用键 key 对应值 value，是查表与计数的核心结构。

当「键可能不存在」时，直接用 `d[k]` 会报错；`get` 与 `setdefault` 能优雅地处理这种情况：

- `get(k, 默认值)`：键不在就回传默认值，不会改动字典。
- `setdefault(k, 默认值)`：键不在就先插入默认值再回传，常用来累加或收集。
- `items()` / `keys()`：分别取出「键值对」与「所有键」来遍历。

````python
# 概念：字典创建、get、items、keys、setdefault
# 造一份商品名称对应价格的字典当示范
prices = {"apple": 30, "banana": 15, "cherry": 80}

# 注意：用 get 安全取值，键不存在时回传默认值而非报错
print("apple 价格：", prices.get("apple", 0))
print("durian 价格：", prices.get("durian", 0))   # 不存在，回传 0

print("所有商品：", list(prices.keys()))

# 概念：用 setdefault 累加统计每个分类出现次数
labels = ["水果", "饮料", "水果", "水果", "饮料"]
counts = {}
for label in labels:
    # setdefault：键第一次出现时先放 0，之后直接 +1
    counts[label] = counts.setdefault(label, 0) + 1

for name, n in counts.items():   # items 同时拿到键与值
    print(f"{name}：{n} 次")
````

## 元组与拆包

元组 tuple 是一串「不可变 immutable」的有序数据，创建后不能添加或修改元素，适合表达固定成组的数据（例如座标）。

拆包 unpacking 让你一次把元组或清单的元素分配给多个变量，使函数回传多值与座标式数据的读写更直觉。

形状：
```
a, b, c = (1, 2, 3)
```

````python
# 概念：元组创建、不可变特性、拆包与多重赋值
# 造一个座标点当示范（tuple 用小括号或逗号即可）
point = (3, 7)
x, y = point                  # 拆包：一次把两个值分给 x 与 y
print("x =", x, "y =", y)

# 注意：tuple 不可变，下面这行若运行会报 TypeError（故仅说明不运行）
# point[0] = 99

# 概念：拆包用于「姓名, 分数」配对
record = ("Alice", 92)
name, score = record
print(f"{name} 得 {score} 分")

# 技巧：拆包可不用暂存变量就交换两个值
a, b = 1, 2
a, b = b, a                   # 右边先打包成 (2, 1) 再拆给左边
print("交换后 a =", a, "b =", b)
````

## 集合与集合运算

集合 set 是一群「不重复」且无顺序的元素，创建时会自动去除重复，并提供很快的成员测试 in。

适合做唯一值整理与关系比较：

| 运算 | 写法 | 意义 |
|---|---|---|
| 联集 union | `a \| b` | 两边全部（去重） |
| 交集 intersection | `a & b` | 两边都有 |
| 差集 difference | `a - b` | 只在 a、不在 b |

````python
# 概念：集合创建、成员测试、add、union、交集与差集
# 造两份修课学生名单当示范（注意 set 会自动去重）
math_class = {"Amy", "Ben", "Cathy", "David", "Amy"}
art_class = {"Ben", "David", "Eric"}

print("数学班（去重后）：", math_class)
print("Amy 有修数学吗？", "Amy" in math_class)   # 成员测试

math_class.add("Frank")       # 添加一名学生

# 概念：联集 / 交集 / 差集做关系比较
print("共同修课（交集）：", math_class & art_class)
print("只修数学（差集）：", math_class - art_class)
print("所有修课学生（联集）：", math_class | art_class)
````

## 推导式

推导式 comprehension 用一行就表达「转换 + 过滤」，取代 for 循环加 append，是 Python 风格的代表写法。

数据量大、逻辑单纯时更短也更易读。

形状：
```
[表达式 for 变量 in 可迭代物 if 条件]
{键: 值 for 变量 in 可迭代物}
```

````python
# 概念：清单推导式、字典推导式、含条件过滤的推导式
# 造一串数字当示范
nums = [1, 2, 3, 4, 5, 6]

# 清单推导式：一行产生每个数的平方
squares = [n * n for n in nums]
print("平方清单：", squares)

# 含条件过滤 filtering：if 放在最后，只留偶数
evens = [n for n in nums if n % 2 == 0]
print("偶数：", evens)

# 字典推导式：创建「数字 -> 其平方」的对应
square_map = {n: n * n for n in nums}
print("数字对平方：", square_map)
````

## 产生器与惰性求值

产生器表达式 generator expression 长得像清单推导式，但用小括号，而且只在「需要下一个值」时才实时计算，这种行为称为惰性求值 lazy evaluation。

好处是处理大量数据时不必一次把整串结果存进内存；代价是它只能逐项取用 iteration 一次，用完就空了。

形状：
```
(表达式 for 变量 in 可迭代物)
```

````python
# 概念：产生器表达式、惰性求值、逐项取用
# 造一个大范围数列当示范（range 本身就是惰性的）
big_range = range(1, 1_000_001)

# 产生器用小括号：此刻并不会真的算出一百万个值
gen = (n * n for n in big_range)

# 注意：sum 会逐项向 gen 索取值并实时累加，全程不创建完整清单
total = sum(gen)
print("前一百万个平方和：", total)

# 注意：产生器用完即空，再取就什么都没有
print("再次 sum：", sum(gen))   # 得到 0，因为 gen 已被耗尽

# 技巧：清单推导式会一次算完并占内存，产生器则边用边算
small_list = [n * n for n in range(5)]   # 立刻就有完整结果可重复使用
print("清单（可重复用）：", small_list)
````

## 迭代辅助工具

`enumerate` 与 `zip` 让循环更干净：

- `enumerate(xs)`：走访时同时拿到「索引 index」与「元素」，不必自己维护计数器。
- `zip(xs, ys)`：把多个串行「并行配对」，每次各取一个组成一组。

两者都常与拆包搭配，直接把配对结果分给多个变量。

````python
# 概念：enumerate 取索引、zip 并行配对、与拆包搭配
# 造名次与名字两串数据当示范
names = ["冠军", "亚军", "季军"]
scores = [98, 95, 91]

# enumerate 预设从 0 起算，start=1 让编号从 1 开始更自然
for rank, name in enumerate(names, start=1):
    print(f"第 {rank} 名：{name}")

# zip 把两串配成对，搭配拆包一次拿到对应的两个值
for name, score in zip(names, scores):
    print(f"{name} 得分 {score}")
````

## 常用内置函数

内置函数 built-in function 不需导入即可使用，这组是处理容器数据的常备工具箱：

| 函数 | 作用 |
|---|---|
| `len` / `sum` | 算数量 / 总和 |
| `min` / `max` | 最小 / 最大 |
| `sorted` | 回传排序后的新清单 |
| `round` / `abs` | 四舍五入 / 绝对值 |
| `isinstance` | 检查类型 type |

````python
# 概念：len、sum、min、max、sorted、round、abs、isinstance
# 造一串分数当示范
scores = [88, 92, 79, 95, 60]

count = len(scores)                  # 笔数
total = sum(scores)                  # 总分
average = total / count

print("总分：", total, "平均：", round(average, 1))   # round 第二参数控制小数字
print("最高：", max(scores), "最低：", min(scores))
print("由高到低排序：", sorted(scores, reverse=True))  # sorted 不改原清单

# abs 取绝对值：与平均的偏差
print("最高分与平均的偏差：", round(abs(max(scores) - average), 1))

# isinstance 检查类型，回传 True / False
print("scores 是 list 吗？", isinstance(scores, list))
````
