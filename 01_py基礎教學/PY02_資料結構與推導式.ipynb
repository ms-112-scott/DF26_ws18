## 清單與索引切片

清單 list 是一串有順序、可以隨時修改的資料，是 Python 最常用的可變序列 mutable sequence。

先打穩元素的新增、定位與區段擷取，後續的推導式才有著力點。常用操作：

| 操作 | 寫法 | 作用 |
|---|---|---|
| 附加 append | `xs.append(v)` | 在尾端加入單一元素 |
| 延伸 extend | `xs.extend(ys)` | 把另一串元素逐一接到尾端 |
| 索引 indexing | `xs[i]` | 取出第 i 個元素（從 0 起算） |
| 切片 slicing | `xs[a:b]` | 取出 a 到 b（不含 b）的區段 |

````python
# 概念：清單建立、append、extend、索引與切片
# 造一週的每日氣溫（攝氏度）當示範資料
temps = [22, 24, 19, 21, 25, 23, 20]

temps.append(26)              # 在尾端補上「第八天」一筆新值
temps.extend([18, 17])        # 一次接上兩天資料（extend 是逐一展開，不是塞入一個 list）

# 注意：索引從 0 起算，負索引從尾端倒數，-1 是最後一個
print("第一天：", temps[0])
print("最後一天：", temps[-1])

# 切片 [a:b] 取 a 到 b-1；省略起點代表從頭，負數從尾端算
print("後三天：", temps[-3:])   # 取最後三筆
print("全部：", temps)
````

## 字典的鍵值操作

字典 dict 用鍵 key 對應值 value，是查表與計數的核心結構。

當「鍵可能不存在」時，直接用 `d[k]` 會報錯；`get` 與 `setdefault` 能優雅地處理這種情況：

- `get(k, 預設值)`：鍵不在就回傳預設值，不會改動字典。
- `setdefault(k, 預設值)`：鍵不在就先插入預設值再回傳，常用來累加或收集。
- `items()` / `keys()`：分別取出「鍵值對」與「所有鍵」來遍歷。

````python
# 概念：字典建立、get、items、keys、setdefault
# 造一份商品名稱對應價格的字典當示範
prices = {"apple": 30, "banana": 15, "cherry": 80}

# 注意：用 get 安全取值，鍵不存在時回傳預設值而非報錯
print("apple 價格：", prices.get("apple", 0))
print("durian 價格：", prices.get("durian", 0))   # 不存在，回傳 0

print("所有商品：", list(prices.keys()))

# 概念：用 setdefault 累加統計每個分類出現次數
labels = ["水果", "飲料", "水果", "水果", "飲料"]
counts = {}
for label in labels:
    # setdefault：鍵第一次出現時先放 0，之後直接 +1
    counts[label] = counts.setdefault(label, 0) + 1

for name, n in counts.items():   # items 同時拿到鍵與值
    print(f"{name}：{n} 次")
````

## 元組與拆包

元組 tuple 是一串「不可變 immutable」的有序資料，建立後不能新增或修改元素，適合表達固定成組的資料（例如座標）。

拆包 unpacking 讓你一次把元組或清單的元素分配給多個變數，使函式回傳多值與座標式資料的讀寫更直覺。

形狀：
```
a, b, c = (1, 2, 3)
```

````python
# 概念：元組建立、不可變特性、拆包與多重賦值
# 造一個座標點當示範（tuple 用小括號或逗號即可）
point = (3, 7)
x, y = point                  # 拆包：一次把兩個值分給 x 與 y
print("x =", x, "y =", y)

# 注意：tuple 不可變，下面這行若執行會報 TypeError（故僅說明不執行）
# point[0] = 99

# 概念：拆包用於「姓名, 分數」配對
record = ("Alice", 92)
name, score = record
print(f"{name} 得 {score} 分")

# 技巧：拆包可不用暫存變數就交換兩個值
a, b = 1, 2
a, b = b, a                   # 右邊先打包成 (2, 1) 再拆給左邊
print("交換後 a =", a, "b =", b)
````

## 集合與集合運算

集合 set 是一群「不重複」且無順序的元素，建立時會自動去除重複，並提供很快的成員測試 in。

適合做唯一值整理與關係比較：

| 運算 | 寫法 | 意義 |
|---|---|---|
| 聯集 union | `a \| b` | 兩邊全部（去重） |
| 交集 intersection | `a & b` | 兩邊都有 |
| 差集 difference | `a - b` | 只在 a、不在 b |

````python
# 概念：集合建立、成員測試、add、union、交集與差集
# 造兩份修課學生名單當示範（注意 set 會自動去重）
math_class = {"Amy", "Ben", "Cathy", "David", "Amy"}
art_class = {"Ben", "David", "Eric"}

print("數學班（去重後）：", math_class)
print("Amy 有修數學嗎？", "Amy" in math_class)   # 成員測試

math_class.add("Frank")       # 新增一名學生

# 概念：聯集 / 交集 / 差集做關係比較
print("共同修課（交集）：", math_class & art_class)
print("只修數學（差集）：", math_class - art_class)
print("所有修課學生（聯集）：", math_class | art_class)
````

## 推導式

推導式 comprehension 用一行就表達「轉換 + 過濾」，取代 for 迴圈加 append，是 Python 風格的代表寫法。

資料量大、邏輯單純時更短也更易讀。

形狀：
```
[運算式 for 變數 in 可迭代物 if 條件]
{鍵: 值 for 變數 in 可迭代物}
```

````python
# 概念：清單推導式、字典推導式、含條件過濾的推導式
# 造一串數字當示範
nums = [1, 2, 3, 4, 5, 6]

# 清單推導式：一行產生每個數的平方
squares = [n * n for n in nums]
print("平方清單：", squares)

# 含條件過濾 filtering：if 放在最後，只留偶數
evens = [n for n in nums if n % 2 == 0]
print("偶數：", evens)

# 字典推導式：建立「數字 -> 其平方」的對應
square_map = {n: n * n for n in nums}
print("數字對平方：", square_map)
````

## 產生器與惰性求值

產生器運算式 generator expression 長得像清單推導式，但用小括號，而且只在「需要下一個值」時才即時計算，這種行為稱為惰性求值 lazy evaluation。

好處是處理大量資料時不必一次把整串結果存進記憶體；代價是它只能逐項取用 iteration 一次，用完就空了。

形狀：
```
(運算式 for 變數 in 可迭代物)
```

````python
# 概念：產生器運算式、惰性求值、逐項取用
# 造一個大範圍數列當示範（range 本身就是惰性的）
big_range = range(1, 1_000_001)

# 產生器用小括號：此刻並不會真的算出一百萬個值
gen = (n * n for n in big_range)

# 注意：sum 會逐項向 gen 索取值並即時累加，全程不建立完整清單
total = sum(gen)
print("前一百萬個平方和：", total)

# 注意：產生器用完即空，再取就什麼都沒有
print("再次 sum：", sum(gen))   # 得到 0，因為 gen 已被耗盡

# 技巧：清單推導式會一次算完並佔記憶體，產生器則邊用邊算
small_list = [n * n for n in range(5)]   # 立刻就有完整結果可重複使用
print("清單（可重複用）：", small_list)
````

## 迭代輔助工具

`enumerate` 與 `zip` 讓迴圈更乾淨：

- `enumerate(xs)`：走訪時同時拿到「索引 index」與「元素」，不必自己維護計數器。
- `zip(xs, ys)`：把多個序列「並行配對」，每次各取一個組成一組。

兩者都常與拆包搭配，直接把配對結果分給多個變數。

````python
# 概念：enumerate 取索引、zip 並行配對、與拆包搭配
# 造名次與名字兩串資料當示範
names = ["冠軍", "亞軍", "季軍"]
scores = [98, 95, 91]

# enumerate 預設從 0 起算，start=1 讓編號從 1 開始更自然
for rank, name in enumerate(names, start=1):
    print(f"第 {rank} 名：{name}")

# zip 把兩串配成對，搭配拆包一次拿到對應的兩個值
for name, score in zip(names, scores):
    print(f"{name} 得分 {score}")
````

## 常用內建函式

內建函式 built-in function 不需匯入即可使用，這組是處理容器資料的常備工具箱：

| 函式 | 作用 |
|---|---|
| `len` / `sum` | 算數量 / 總和 |
| `min` / `max` | 最小 / 最大 |
| `sorted` | 回傳排序後的新清單 |
| `round` / `abs` | 四捨五入 / 絕對值 |
| `isinstance` | 檢查型別 type |

````python
# 概念：len、sum、min、max、sorted、round、abs、isinstance
# 造一串分數當示範
scores = [88, 92, 79, 95, 60]

count = len(scores)                  # 筆數
total = sum(scores)                  # 總分
average = total / count

print("總分：", total, "平均：", round(average, 1))   # round 第二參數控制小數位
print("最高：", max(scores), "最低：", min(scores))
print("由高到低排序：", sorted(scores, reverse=True))  # sorted 不改原清單

# abs 取絕對值：與平均的偏差
print("最高分與平均的偏差：", round(abs(max(scores) - average), 1))

# isinstance 檢查型別，回傳 True / False
print("scores 是 list 嗎？", isinstance(scores, list))
````
